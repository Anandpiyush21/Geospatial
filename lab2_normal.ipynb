```python
from PIL import Image  # Library to open and work with images
import numpy as np  # Library for array operations
import matplotlib.pyplot as plt  # Library to display images

# Step 1: Open the image
image_path = 'download.jpeg'  # Path to your JPEG file
image = Image.open(image_path)

# Step 2: Convert the image to black and white (grayscale)
bw_image = image.convert('L')  # 'L' mode is for grayscale

# Step 3: Convert the black and white image into a numpy array
image_matrix = np.array(bw_image)

# Step 4: Get image dimensions (height and width)
height = image_matrix.shape[0]
width = image_matrix.shape[1]

# Step 5: Define the coordinates to segment the image into 4 parts (2x2 grid)
half_height = height // 2
half_width = width // 2

# Slice the image into 4 quadrants:
# Quadrant 1 (Top-left)
quad_1 = image_matrix[:half_height, :half_width]
# Quadrant 2 (Top-right)
quad_2 = image_matrix[:half_height, half_width:]
# Quadrant 3 (Bottom-left)
quad_3 = image_matrix[half_height:, :half_width]
# Quadrant 4 (Bottom-right)
quad_4 = image_matrix[half_height:, half_width:]

# Step 6: Display each quadrant
plt.figure(figsize=(10, 8))

# Plotting Quadrant 1
plt.subplot(2, 2, 1)
plt.imshow(quad_1, cmap='gray')
plt.title("Quadrant 1 - Top Left")
plt.axis('off')

# Plotting Quadrant 2
plt.subplot(2, 2, 2)
plt.imshow(quad_2, cmap='gray')
plt.title("Quadrant 2 - Top Right")
plt.axis('off')

# Plotting Quadrant 3
plt.subplot(2, 2, 3)
plt.imshow(quad_3, cmap='gray')
plt.title("Quadrant 3 - Bottom Left")
plt.axis('off')

# Plotting Quadrant 4
plt.subplot(2, 2, 4)
plt.imshow(quad_4, cmap='gray')
plt.title("Quadrant 4 - Bottom Right")
plt.axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
from PIL import Image  # Library to open and work with images
import numpy as np  # Library for array operations
import matplotlib.pyplot as plt  # Library to display images

# Step 1: Open the image
image_path = 'download.jpeg'  # Path to your JPEG file
image = Image.open(image_path)

# Step 2: Convert the image to black and white (grayscale)
bw_image = image.convert('L')  # 'L' mode is for grayscale

# Step 3: Convert the black and white image into a numpy array
image_matrix = np.array(bw_image)

# Step 4: Apply thresholding
threshold_value = 128  # You can adjust this value
thresholded_image = np.where(image_matrix > threshold_value, 255, 0)

# Step 5: Display the thresholded image
plt.imshow(thresholded_image, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')  # Hide axes for clarity
plt.show()
```

```python
from PIL import Image  # Library to open and work with images
import numpy as np  # Library for array operations

# Step 1: Open the image
image_path = 'download.jpeg'  # Path to your JPEG file
image = Image.open(image_path)

# Step 2: Convert the image to black and white (grayscale)
bw_image = image.convert('L')  # 'L' mode is for grayscale

# Step 3: Convert the black and white image into a numpy array
image_matrix = np.array(bw_image)

# --- Min-Max Normalization (Scale to range [0, 1]) ---
min_val = image_matrix.min()
max_val = image_matrix.max()
min_max_normalized = (image_matrix - min_val) / (max_val - min_val)

# Step 4: Show the normalized image
import matplotlib.pyplot as plt
plt.imshow(min_max_normalized, cmap='gray')
plt.title("Min-Max Normalized Image")
plt.axis('off')  # Hide axes for clarity
plt.show()
```
