```python
import rasterio  # Library to work with .tif files
import matplotlib.pyplot as plt  # Library to create plots

# Step 1: Specify the file path
file_path = "NextGen_Tokyo_JP_15m_Geo.tif"

# Step 2: Open the file using rasterio
with rasterio.open(file_path) as file:
    # Step 3: Get the total number of bands in the file
    total_bands = file.count
    print("Total number of bands:", total_bands)
    
    # Step 4: Create a figure to display the subplots
    fig, axes = plt.subplots(1, total_bands, figsize=(15, 5))  # Adjust figsize to fit all subplots
    
    # Step 5: Loop through each band and display it in the corresponding subplot
    for band_number in range(1, total_bands + 1):  # Bands start from 1, not 0
        # Read the band data
        band_data = file.read(band_number)
        
        # Plot the band in the corresponding subplot
        ax = axes[band_number - 1]  # Get the correct subplot axis
        im = ax.imshow(band_data, cmap='brg')  # Display the band with the brg color map
        ax.set_title(f"Band {band_number}")  # Add a title for the subplot
        ax.axis('off')  # Turn off the axis for better presentation
        
        # Add colorbar to the last subplot
        if band_number == total_bands:
            fig.colorbar(im, ax=ax, orientation='vertical')

    # Step 6: Show the figure with all subplots
    plt.tight_layout()  # Adjust layout to avoid overlap
    plt.show()
```

```python
import rasterio  # Library to work with .tif files
import matplotlib.pyplot as plt  # Library to create plots
import numpy as np  # Library for array operations

# Step 1: Specify the file path
file_path = "NextGen_Tokyo_JP_15m_Geo.tif"

# Step 2: Open the file using rasterio
with rasterio.open(file_path) as file:
    # Step 3: Read the Red, Green, and Blue bands
    band_1 = file.read(1)  # Blue (Band 1)
    band_2 = file.read(2)  # Green (Band 2)
    band_3 = file.read(3)  # Red (Band 3)

    # Step 4: Create True Color Composite (TCC) by stacking the bands
    tcc = np.stack([band_3, band_2, band_1], axis=-1)  # RGB (Red, Green, Blue)

    # Step 5: Display the TCC image
    plt.imshow(tcc)
    plt.title("True Color Composite (TCC)")
    plt.axis('off')  # Turn off axis for better presentation
    plt.show()
```

```python
import rasterio  # Library to work with .tif files
import rasterio.plot  # Library to plot raster images
import matplotlib.pyplot as plt  # Library to create plots

# Step 1: Open the file
data_name = r"NextGen_Tokyo_JP_15m_Geo.tif"
tiff = rasterio.open(data_name)

# Step 2: Get the image dimensions (height and width)
height = tiff.height
width = tiff.width

# Step 3: Define the coordinates to segment the image into 4 parts (2x2 grid)
half_height = height // 2
half_width = width // 2

# Slice the image into 4 quadrants:
# Quadrant 1 (Top-left)
quad_1 = tiff.read(1)[:, :half_width]
# Quadrant 2 (Top-right)
quad_2 = tiff.read(1)[:, half_width:]
# Quadrant 3 (Bottom-left)
quad_3 = tiff.read(1)[half_height:, :half_width]
# Quadrant 4 (Bottom-right)
quad_4 = tiff.read(1)[half_height:, half_width:]

# Step 4: Display each quadrant

# Plotting Quadrant 1
plt.figure(figsize=(8, 6))
plt.subplot(2, 2, 1)
plt.imshow(quad_1, cmap='gist_heat')
plt.title("Quadrant 1 - Top Left")
plt.axis('off')

# Plotting Quadrant 2
plt.subplot(2, 2, 2)
plt.imshow(quad_2, cmap='gist_heat')
plt.title("Quadrant 2 - Top Right")
plt.axis('off')

# Plotting Quadrant 3
plt.subplot(2, 2, 3)
plt.imshow(quad_3, cmap='gist_heat')
plt.title("Quadrant 3 - Bottom Left")
plt.axis('off')

# Plotting Quadrant 4
plt.subplot(2, 2, 4)
plt.imshow(quad_4, cmap='gist_heat')
plt.title("Quadrant 4 - Bottom Right")
plt.axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```
