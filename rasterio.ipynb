```python
import rasterio
from rasterio.windows import from_bounds
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Open the GeoTIFF file
tiff = rasterio.open(r"C:\Users\anand\Code_basics\GDA\NextGen_Tokyo_JP_15m_Geo\NextGen_Tokyo_JP_15m_Geo.tif")

# Define the latitude and longitude bounds (example values)
min_lon, min_lat = 139.5, 35.0  # Bottom-left corner (longitude, latitude)
max_lon, max_lat = 140.0, 35.5  # Top-right corner (longitude, latitude)

# Transform the latitude and longitude bounds to image coordinates
window = from_bounds(min_lon, min_lat, max_lon, max_lat, tiff.transform)

# Read the subset of the image
subset = tiff.read(1, window=window)

# Set up the plot with Cartopy's PlateCarree projection
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})

# Plot the subset of the image
ax.imshow(subset, extent=(min_lon, max_lon, min_lat, max_lat), transform=ccrs.PlateCarree(), cmap='Greys')

# Add gridlines
gridlines = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
gridlines.top_labels = False
gridlines.right_labels = False

# Add land and ocean features for better context
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)

# Set up the latitude and longitude labels
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# Show the plot
plt.show()
```
